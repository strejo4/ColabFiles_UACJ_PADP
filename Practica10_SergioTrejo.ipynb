## Histogramas con plotly y el dataset tips de Seaborn

Link repositorio github: https://github.com/strejo4/ColabFiles_UACJ_PADP/blob/main/Practica10_SergioTrejo.ipynb

Objetivo: Analizar la distribución de las propinas (tip) con histogramas interactivos.

INSTRUCCIONES

1. Cargar el dataset tips desde seaborn. (sns.load_dataset("tips"))
2. Usar plotly para graficar un histograma de las propinas (tip).
3. Personalizar el número de bins y agregar colores según el día de la semana (day).

```python
#Cargar el dataset tips desde seaborn

import seaborn as sns
import plotly.express as px

df = sns.load_dataset("tips")
df.head()
```

```python
#Usar plotly  para graficar un histograma de las propinas (tip)

fig = px.histogram(df, x='tip',  labels= {'tip':'Propina', 'day':'Dia de la semana'}, title='Distribucion de las propinas')
fig.update_layout(bargap=0.2)
fig.show()
```

```python
#Personalizar el número de bins y agregar colores según el día de la semana (day).

fig = px.histogram(df, x='tip', nbins=30, color="day", labels= {'tip':'Propina', 'day':'Dia de la semana'}, title='Distribucion de las propinas')
fig.update_layout(bargap=0.2)
fig.show()
```

# Datos importantes:

* El nombre de la accion y su ticker (por ejemplo, Cisco Systems Inc. (CSCO), Amazon.com Inc. (AMZN)).

* La fecha y hora de la medicion de la accion (por ejemplo, Fecha: 2024-09-02 12:34:26).

* El valor actual (por ejemplo, Valor actual: $446.52).

Estos elementos es donde podemos extraer informacion clave ya que nos dan los datos que realmente describen el estado de cada accion en el momento del informe o de la notificacion por correo electronico.

Usa re para extraer la siguiente información:
  * Nombre de la empresa: Busca el patrón del nombre en las líneas del asunto.
  * Fecha del valor de la acción: Encuentra el formato de fecha.
  * Valor de la acción: Extrae números después de "Valor actual:".

Análisis Estadístico:
  * Para cada empresa extraída:
    * Encuentra el valor máximo, mínimo, promedio


```python
# Codigo para extraer los valores de todas las empresas y mostrar estadisticos
import re
# Abrir el archivo
informe = open('/content/drive/My Drive/ClassFiles/informe_acciones.txt', encoding='latin-1')

# Diccionario para guardar datos por empresa
datos_empresas = {}

# Recorrer cada línea del archivo
for linea in informe:
    linea = linea.rstrip()

    # Buscar nombre y ticker de la empresa
    nombre_ticker = re.findall(r'acción de (.+)\s\((.+)\)', linea) # Busca "accion de ", (.+) = nombre de la empresa, \s = espacio, \( y \) = parentesis, (.+) = ticker
    if len(nombre_ticker) > 0:
        nombre, ticker = nombre_ticker[0]
        empresa = f"{nombre} ({ticker})"
        if empresa not in datos_empresas:
            datos_empresas[empresa] = {'fechas': [], 'valores': []}

    # Buscar fecha
    fecha = re.findall(r'^Fecha: (.+)', linea) # Busca "^" inicio de la linea, "Fecha: ", (.+) = el resto de la linea
    if len(fecha) > 0:
        datos_empresas[empresa]['fechas'].append(fecha[0])

    # Buscar valor actual
    valor = re.findall(r'^Valor actual: \$?([0-9.]+)', linea) # Busca "^" inicio de la linea, "Valor actual: ", \$? = signo $ opcional, ([0-9.]+) = numeros y puntos
    if len(valor) > 0:
        datos_empresas[empresa]['valores'].append(float(valor[0]))

# Realizar analisis estadistico e Imprimir tabla con el formato deseado
print(f"{'Empresa':<35}{'Fecha mas reciente':<20}{'Valor Max ($)':<15}{'Valor Min ($)':<15}{'Promedio ($)':<15}'")
for empresa, info in datos_empresas.items():
    if len(info['valores']) > 0:
        fecha_reciente = max(info['fechas'])
        valor_max = max(info['valores'])
        valor_min = min(info['valores'])
        promedio = sum(info['valores']) / len(info['valores'])
        print(f'{empresa:<35}{fecha_reciente:<20}{valor_max:<15.2f}{valor_min:<15.2f}{promedio:<15.2f}')




```

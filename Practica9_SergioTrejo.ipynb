## Gráficos de dispersión con plotly.express y el dataset gapminder

Link repositorio github: https://github.com/strejo4/ColabFiles_UACJ_PADP/blob/main/Practica9_SergioTrejo.ipynb

Objetivo: Representar la relación entre el PIB per cápita y la esperanza de vida de distintos países usando plotly.

Instrucciones:

1. Cargar el dataset gapminder de plotly.express, puede emplear las siguientes expresiones.
import plotly.express as px

df = px.data.gapminder()

2. Filtrar los datos para el año 2007.

3. Crear un gráfico de dispersión donde el eje X sea el PIB per cápita (gdpPercap) y el eje Y la esperanza de vida (lifeExp).

4. Usar el tamaño de los puntos para representar la población (pop).
Colorear los puntos según el continente.

```python
#Importar la libreria plotly.express como px
#Cargar el data set gapminder

import plotly.express as px
df = px.data.gapminder()
df.head()
```

```python
#Filtar los datos para el año 2007.
filtro_2007 = df[df['year'] == 2007].copy()
filtro_2007.shape, filtro_2007.head() #despues del filtrar el DF quedo con 142 lines y 8 columnas.
```

```python
#Crear un gráfico de dispersión donde el eje X sea el PIB per cápita (gdpPercap) y el eje Y la esperanza de vida (lifeExp).
fig = px.scatter(filtro_2007, x = 'gdpPercap', y = 'lifeExp',labels={"gdpPercap":"PIB per cápita","lifeExp":"Esperanza de vida"}, color = 'continent', title = 'Esperanza de Vida Vs PIB Per Capita')
fig.show()

```

```python
#Usar el tamaño de los puntos para representar la población (pop). Colorear los puntos según el continente.
fig = px.scatter(filtro_2007, x = 'gdpPercap', y = 'lifeExp',labels={"gdpPercap":"PIB per cápita","lifeExp":"Esperanza de vida"}, size = 'pop', color = 'continent', title = 'Esperanza de Vida Vs PIB Per Capita')
fig.show()
```

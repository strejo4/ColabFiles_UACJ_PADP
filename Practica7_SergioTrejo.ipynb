**Crear los siguientes gráficos con Matplotlib:**

a) Gráfico de barras:

* Objetivo: Comparar el promedio de largo y ancho de los pétalos entre las tres especies de Iris.
* Instrucciones: Utiliza un gráfico de barras para mostrar la media de las columnas petal_length y petal_width agrupadas por species.

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Cargard los datos
df = sns.load_dataset('iris')
print(df.head())
print(df.info())


#Caluclar el promedio de largo y ancho de petalos agrupados por especie
promedio_petalos = df.groupby('species')[['petal_length', 'petal_width']].mean()
print(promedio_petalos)

#Grafica de barras
promedio_petalos.plot(kind='bar', color=['blue','red'])
plt.xlabel('Especie')
plt.ylabel('Promedio')
plt.title('Promedio de largo y ancho de pétalos por especie')
plt.show()

# Segun los datos, se observa que la especie setosa tiene petalos mas pequenos en promedio, mientras que virginica presenta los petalos mas largos y anchos.
```

**b) Histograma:**

* Objetivo: Visualizar la distribución de los largos de los sépalos (sepal_length).
* Instrucciones: Crea un histograma de la variable sepal_length y asegúrate de que tenga un número adecuado de bins. Coloca etiquetas en los ejes y un título

```python
#Crear histograma de la columna 'sepal_lenght'
plt.hist(df['sepal_length'], bins=10, color='green', edgecolor='black')
plt.title('Distribucion de Sepalos')
plt.xlabel('Largo del sepalo')
plt.ylabel('Frecuencia')
plt.show()

#El histograma muestra como se distribuyen los valores del largo del sepalo en el dataset,
#El histogram nos puede ayudar a interpretar como una aproxicamecion de la funcion de densidad de probabilidad.
```

c) Gráfico de dispersión (scatter plot):

* Objetivo: Relacionar el largo del pétalo con el ancho del pétalo.
* Instrucciones: Crea un gráfico de dispersión que compare petal_length con petal_width. Usa un color diferente para cada especie.

```python
#Grafica scatter plot o de dispersion
plt.scatter(df['petal_length'],df['petal_width'], alpha = 0.5, c=df['species'].astype('category').cat.codes, cmap='coolwarm')
# astype convierte species a tipo categoria y .cat le da valores numericos para que cada especie tenga su color distinto con el mapa de color coolwarm.
# Alpha controla la transparencia de los puntos en el scatter plot para qe no se pierda o quede empalmado (0 invisible, 1 opaco, 0.5 semi-transparente).
plt.title('Relacion largo y ancho de los petalos')
plt.xlabel('Largo petalos')
plt.ylabel('Ancho petalos')
plt.show()

# El plot de dispersion muestra como se consentran los datos
# A mayor largo del petalo tiende a haber mayor ancho del petalo, mostrando una relacion entre ambas variables.
```

Gráfico de cajas (box plot):

* Objetivo: Analizar la distribución de las medidas del sépalo (sepal_length, sepal_width).
* Instrucciones: Crea un gráfico de cajas para mostrar la distribución de las variables sepal_length y sepal_width por especie.

```python
# Graficos de cajas o box plot

#Separar los datos por especie y 'sepal_length'
setosa = df[df['species'] == 'setosa']['sepal_length'] # Filtrar las filas donde species es 'setosa' y seleccionar solo la columna 'sepal_length'
versicolor = df[df['species'] == 'versicolor']['sepal_length']
virginica = df[df['species'] == 'virginica']['sepal_length']

#Separar los datos por especie y 'sepal_width'
setosa2 = df[df['species'] == 'setosa']['sepal_width']
versicolor2 = df[df['species'] == 'versicolor']['sepal_width']
virginica2 = df[df['species'] == 'virginica']['sepal_width']


##Box plot sepal_length
plt.boxplot([setosa, versicolor, virginica], labels =['Setosa', 'Versicolor', 'Virginica'])
plt.title('Distribucion del largo de los Sepalos por especie')
plt.xlabel('Especie')
plt.ylabel('Largo sepalo')
plt.show()

#Box plot sepal_width
plt.boxplot([setosa2, versicolor2, virginica2], labels =['Setosa', 'Versicolor', 'Virginica'])
plt.title('Distribucion del ancho de los Sepalos por especie')
plt.xlabel('Especie')
plt.ylabel('Ancho sepalo')
plt.show()


# Los boxplots muestra diferencias claras entre especies y tambien outliers (puntos fuera de los bigotes) que indican valores atipicos en algunas flores.
```

e) Gráfico de líneas:

* Objetivo: Mostrar la tendencia de la longitud del pétalo (petal_length) a través de las observaciones.
* Instrucciones: Crea un gráfico de líneas utilizando la variable petal_length (en el eje Y) con el índice de las observaciones (en el eje X). Puedes hacer esto para ver cómo varía la longitud del pétalo a lo largo del dataset.

```python
# Grafico de lineas para petal_length
plt.plot(df.index, df['petal_length'], color='blue')
plt.title('Tendencia del largo de los pétalos a traves de las observaciones')
plt.xlabel('Indice de observaciones')
plt.ylabel('Largo del petalo (petal_length)')
plt.show()

## Este grafico de lineas permite observar la variacion del largo de los petalos a lo largo del dataset y detectar patrones o saltos notables.
```

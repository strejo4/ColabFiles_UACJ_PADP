## 1. Operaciones basicas
**Objetivo: Crear y manipular arrays.
Instrucciones:**
1. Crea un array de 10 elementos con valores entre 1 y 10.
2. Calcula la media, mediana, y desviacion estandar de los valores.
3. Multiplica cada valor por 2.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 1.Crear un array de 10 elementos con valores entre 1 y 10
array = np.random.randint(1, 11, size=10)
print("1.Crear un array de 10 elementos con valores entre 1 y 10")
print("Arreglo de 10 numeros con valores entre 1 y 10: ", array)

# 2.Calcula la media, mediana, y desviacion estandard de los valores
media = np.mean(array)
mediana = np.median(array)
desviacion_estandar = np.std(array)
print("\n2.Calcula la media, mediana, y desviacion estandard de los valores")
print("La media del arreglo es:", media)
print("La mediana del arreglo es:", mediana)
print("La desviacion estandar del arreglo es:", desviacion_estandar)

# 3.Multiplica cada valor por 2
array_multiplicado = array * 2
print("\n3.Multiplica cada valor por 2")
print("Arreglo multiplicado por 2:", array_multiplicado)
```

## 2. Operaciones con DataFrames
**Objetivo: Crear y manipular un DataFrame**

1. Crea un DataFrame con 3 columnas (A, B, C) y 5 filas con numeros
aleatorios enteros.
2. Renombra las columnas a Col1, Col2, Col3.
3. Anade una nueva columna que sea la suma de las columnas Col1 y Col2.
4. Elimina todas las filas donde el valor de Col3 sea menor que 10.

```python
# 1.Crear un DF con 3 columnas (A,B,C) y 5 filas con numeros aleatorios enteros.
df = pd.DataFrame(np.random.randint(0,100, size=(5,3)), columns=['A','B','C'])
print("\n1.Crear un DF con 3 columnas (A,B,C) y 5 filas con numeros aleatorios enteros", df)

# 2.Renombra las columnas a Col1, Col2, Col3.
df.rename(columns={'A':'Col1', 'B':'Col2', 'C': 'Col3'}, inplace=True)
print("\n2. Renombra las columnas a Col1, Col2, Col3.", df)

# 3.Aadir una nueva columna que sea la suma de las columnas Col1 y Col2.
df['Col4'] = df['Col1'] + df['Col2']
print("\n3.Aadir una nueva columna que sea la suma de las columnas Col1 y Col2", df)

# 4.Eliminar todas las filas donde el valor de Col3 sea menor que 10
df = df.drop(df[df['Col3'] < 10].index)
print("\n4. Eliminar todas las filas donde el valor de Col3 sea menor que 10",df)
```

## 3. Estadistica descriptiva
**Aplicar analisis estadstico basico.**

1. Crea un DataFrame con 100 valores aleatorios entre 0 y 50.
2. Calcula la media, mediana y moda de los valores.
3. Encuentra el valor mnimo y el valor maximo del DataFrame.


```python
# 1.Crear un Dataframe con 100 valores aleatorios entre 0 y 50
df = pd.DataFrame(np.random.randint(0,51, size=(100)), columns=['Valor'])
print(df.head())
print("\n1.Crear un Dataframe con 100 valores aleatorios entre 0 y 50\n", df)

# 2.Calcula la media, mediana y moda de los valores
media = df['Valor'].mean()
mediana = df['Valor'].median()
moda = df['Valor'].mode()[0]
print("\n2.Calcula la media, mediana y moda de los valores")
print("La media es:", media)
print("La mediana es:", mediana)
print("La moda es:", moda)

# 3.Encontrar el valor minimo y el valor maximo del DF
valor_minimo = df['Valor'].min()
valor_maximo = df['Valor'].max()
print("\n3.Encontrar el valor minimo y el valor maximo del DF")
print("El valor minimo es:", valor_minimo)
print("El valor maximo es:", valor_maximo)
```

## 4. Analisis de series temporales

**Trabajar con fechas y series temporales.**

1. Crea un DataFrame con una columna de fechas (usa pd.date_range()
para generar las fechas).
2. Genera 100 valores aleatorios asociados a estas fechas.
3. Calcula la media movil de los valores con una ventana de 5 das.

```python
# 1.Crea un DataFrame con una columna de fechas (usa pd.date_range() para generar las fechas).
df = pd.DataFrame({'Fecha': pd.date_range(start='2023-01-01', periods=100)})
print("\n1.Crea un DataFrame con una columna de fechas (usa pd.date_range() para generar las fechas).")
print(df.head())

# 2.Genera 100 valores aleatorios asociados a estas fechas.
df['Valor'] = np.random.randint(0,100, size=100)
print("\n2.Genera 100 valores aleatorios asociados a estas fechas.")
print(df.head())

# 3.Calcula la media movil de los valores con una ventana de 5 dias.
df['Media Movil'] = df['Valor'].rolling(window=5).mean()
print("\n3.Calcula la media movil de los valores con una ventana de 5 dias.")
print(df.head(10)) # Se utilizan 10 dias para mostrar las primeras filas y observar los valores NaN generados antes del calculo de la primera media movil
```

## Grafico de Lneas
**Visualizar la tendencia de una serie de datos.**

1. Crea una serie de datos aleatorios (entre 0 y 100) de tamano 50.
2. Usa Matplotlib para graficar los datos en un grafico de lneas.
3. Anade ttulo, etiquetas a los ejes y una leyenda.
Cambia el color de la lnea a rojo y usa un estilo de lnea punteado.

```python
# 1.Crea una serie de datos aleatorios (entre 0 y 100) de tamano 50.
datos = np.random.randint(0,101, size=50)

# 2.Usa Matplotlib para graficar los datos en un grafico de lineas.
plt.figure(figsize=(10,6))
plt.plot(datos, color='blue', linestyle='-', label='Serie de Datos')
plt.show()

# 3.Anade ttulo, etiquetas a los ejes y una leyenda. Cambia el color de la lnea a rojo y usa un estilo de lnea punteado.
plt.figure(figsize=(10,6))
plt.plot(datos, color='red', linestyle='--', label='Serie de Datos')
plt.title('Grafico de Lineas')
plt.xlabel('X')
plt.ylabel('Y', rotation=0)
plt.legend()
plt.grid(True)
plt.show()
```

Ejercicio 01:Tipo de Datos Correcto

1. Carga el conjunto de datos airdata.csv y verifica que cada columna tenga el tipo de datos correcto
2. Si una columna de tipo temporal no está en el tipo de correcto realiza la conversión.
3. Usa dtypes para mostrar que cada columna, principalmente la que contiene datos temporales, está en el tipo de datos correcto
4. Imprime las primeras tres líneas

```python
import pandas as pd
#Descargar los datos airdata.csv
df = pd.read_csv("/content/drive/MyDrive/ClassFiles/Airdata.csv")
print(df.dtypes)
#Si una columna de tipo temporal no está en el tipo de correcto realiza la conversión.
df['DateTime'] = pd.to_datetime(df['DateTime'], errors='coerce')
#Mostrar los tipos de datos despues de la conversion. Imprimir las primeras 3 lineas
print("\nTipos de datos despues de la conversion:\n")
print(df.dtypes)
print("\nPrimeras 3 lineas:\n")
print(df.head(3))
```

Ejercicio 02: Extraer componentes de fecha y tiempo

1. De la columna 'DateTime' extrae el año, mes, día y hora. Crea una nueva columna para cada componente que extraes
2. Imprime las primeras tres líneas del dataframe que ahora deberá tener 4 nuevas columnas.

```python
#Extraer el año, mes, dia y hora. Crear una nueva columna para cada componente.
df['Year'] = df['DateTime'].dt.year
df['Month'] = df['DateTime'].dt.month
df['Day'] = df['DateTime'].dt.day
df['Hour'] = df['DateTime'].dt.hour
#Imprmir las primeras tres lineas del dataframe
df.head()
```

Ejercicio 03: Aritmética - Suma de Intervalos (timedelta)

1. Suma 10 días, 7 horas y 15 minutos a cada elemento de la columna 'DateTime'
2. Guarda el resultado en una nueva columna y agrega al dataframe.
3. Imprime la columna 'DateTime' y la columna que resulta de sumar 10 días, 7 horas y 15 minutos.

```python
import datetime as dt
#Sumar 10 dias, 7 horas y 15 minutos a cada elemento de la columna 'DateTime"
intervalo = dt.timedelta(days=10, hours=7, minutes=15)
#Columna nueva donde se registran los resultados y se agrega al Data Frame
df['DateTime_Suma'] = df['DateTime'] + intervalo
#Imprimir las columnas DateTime y DateTime_Suma
print(df[['DateTime','DateTime_Suma']].head())
```

## Ejercicio 04: Timestamp

1. Convierte la columna 'DateTime' la cual tiene un formato de fecha legible a a timestamp Unix.
2. Guarda el resultado en una nueva columna
3. Imprime las primeras tres filas de  la columna 'DateTime' y la columna que representa dicha fecha en timestamp Unix

```python
#Convertir la columna DateTime de fecha legible a timestamp Unix
timestamps = []
for fecha in df['DateTime']:
  timestamps.append(fecha.timestamp())
#Gaurdar resultado en una nueva columna
df['DateTime_Unix'] = timestamps
#Imprimir las columnas DateTime y DateTime_Unix. Primeras 3 lineas.
print(df[['DateTime','DateTime_Unix']].head(3))
```

Ejercicio 05: Filtrado de Fechas

1. Utiliza la columna 'DateTime' y Filtra aquellas fechas que son mayores a 2020-11-01

```python
#Filtrar fechas que son mayores a 2020-11-01
df_filtrado = df[df['DateTime'] > '2020-11-01']
#Imprimir el resultado
df_filtrado
```

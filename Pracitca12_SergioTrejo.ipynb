Objetivo

Usar la biblioteca plotly.graph_objects para visualizar de manera interactiva un mapa de calor a partir de un dataset cargado desde seaborn.

Instrucciones:

1. Cargar el dataset "flights" del paquete seaborn.
2. Transforma los datos: Organizar los datos para crear un mapa de calor.
3. Reorganizalos para que los años estén en filas y los meses en columnas.
4. Crear el mapa de calor: Usa el código para crear un mapa de calor con plotly.
5. Personalizar el gráfico: Cambia el título, los colores o agregar otras etiquetas para hacer el gráfico más claro.

```python
# Cargar el dataset "flights" del paquete de seaborn

import seaborn as sns
import plotly.graph_objects as po

df = sns.load_dataset('flights')
df.head()
```

```python
# Transformar los datos: Organizar los datos para crear un mapa de calor. Reorganizalos para que los años estén en filas y los meses en columnas.
#Podemos utilizar el metodo pivot para reaorganizar las lineas y columnas, en index='year, columnas='month', values='passengers

datos_transformados = df.pivot(index='year', columns='month', values='passengers')
datos_transformados.head()
```

```python
# Crear el mapa de calor. Usa el codigo para crear el mapa de calor con plotly

fig = po.Figure(data=po.Heatmap(z=datos_transformados.values, x=datos_transformados.columns, y=datos_transformados.index))
fig.update_layout(title='Mapa de calor de pasajeros por mes y año', xaxis_title = 'Mes', yaxis_title = 'Año')

fig.show()
```

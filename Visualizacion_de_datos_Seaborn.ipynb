```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#Dsitribuciones de las edades de los pasajeros
sns.histplot(titanic['age'], kde=True, color='blue',bins=20) #bins depende de los rangos que queramos determinar por genero.
plt.title('Distribuci√≥n de las edades de los pasajeros')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#Distribucion grafica de barras

sns.barplot(x='sex',y='survived', data=titanic,palette='pastel')
plt.title('Tasa de supervivencia por sexo')
plt.xlabel('Genero')
plt.ylabel('Tasa de supervivencia')
plt.show()
```

```python
#Grafica por categorias para identificar la clase de pasajeros, si sobrevivio o no y que cuente cuantos pasajeros de esa clase sobrevivieron y que muestre cuansto sobrevivieron por cada una de esas clases.

sns.catplot(x='pclass', hue='survived', kind = 'count', data=titanic,palette='Set2')
plt.title('Supervivencia por clase de pasajero')
plt.xlabel('Clase de pasajero')
plt.ylabel('Cantidad de pasajeros')
plt.show()

#Delas 3 clases del data set se distribuyen las clases de pasajeros que sobrevivio o no. Y observar dependiendo el tipo de clase si el pasajero pudo sobrevivir o no.
```

```python
#Box plots. Entender la dispersion de las edades y los outliers de los datos (datos atipicos)
sns.boxplot(x='pclass',y='age', data=titanic,palette='coolwarm')
plt.title('Distribucion de las edades por clase de pasajero')
plt.xlabel('Clase de pasajero')
plt.ylabel('Edad')
plt.show()
#En este caso podemos observar como se ecuentra la distribucion por clase. Fuera de los limites delos bigotes se encuentran los valores atipicos, son dastos que salen afuera de la mediana de los datos.
#Mas adelante veremos una interpretacion durante el analisis de datos
```

```python
#Grafica del mapa de calor. Sirve para identificar la corelacion entre las variables dentro de nuestro dataset. Si observamos el data set original tendremos variables de tipo numericas
#Pero el genero es una cade de caracteres o string.
#Tenemos que identificar variable numericas y para las que no son hay que quitarlas.

num_col = titanic.select_dtypes(include=['number']).columns
corr = titanic[num_col].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm',fmt ='.2f',linewidth=0.5)
plt.title('Mapa de calor de correlacion')
plt.show()

#Esto indica la relacion que existe entre las variables. Identifica si el numero es mas cercano al 1 hay mayor relacion entre si. Si el dato va a color azul, las variables tienen menos relacion entre si.
```

```python
#Greafical Pair Plot . Nos va a servir par averificar como se distribuyen las variables y se relacionan entre si.

sns.pairplot(titanic[['age','fare','pclass', 'survived']])
plt.suptitle('Pairplot de las variables de Titanic')
plt.show()

#Podemos ver la distribucion de diferentes elementos. Nos dan las densidaes y las distribuciones.
```

```python
#Grafica de violin. Nos va aservir para combinar un box plot con una esticamcion de una funcion de densidad de probabilidad.
#Esto no solo nos va ayudar a visualizar la distribucion de los datos, sino tambien com ose forma esa distribucion
#La grafica de violin te vaa yudar para entender la variabilidad y la densidad de los datos.

sns.violinplot(x= 'pclass', y='age',data=titanic,palette='muted')
plt.title('Violin plot de la edad por clase de pasajero')
plt.xlabel('Clase de pasajero')
plt.ylabel('Edad')
plt.show()

#Similar al box plot y una densidad de probabilidad.
```

```python
#Grafica de dispersion individual. El grafico de scatterplot nos va servir para ver relacion entre variables
sns.scatterplot(x='age',y='fare',data=titanic,hue='survived',palette='coolwarm')
plt.title('Relacion entre edad y Tarifa de supervivencia')
plt.xlabel('Edad')
plt.ylabel('Tarifa')
plt.show()

#Grfica para encontrar patrones o relaciones entre dos variables.
```

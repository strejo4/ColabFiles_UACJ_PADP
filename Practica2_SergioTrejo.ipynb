**Diseñar un código en Google Colab que permita a los estudiantes cargar el archivo informe_acciones.txt y extraer información clave utilizando expresiones regulares en Python.**

# Datos importantes:

* El nombre de la accion y su ticker (por ejemplo, Cisco Systems Inc. (CSCO), Amazon.com Inc. (AMZN)).

* La fecha y hora de la medicion de la accion (por ejemplo, Fecha: 2024-09-02 12:34:26).

* El valor actual (por ejemplo, Valor actual: $446.52).

Estos elementos es donde podemos extraer informacion clave ya que nos dan los datos que realmente describen el estado de cada accion en el momento del informe o de la notificacion por correo electronico.


```python
#Codigo para extraer todos los valores de Nombre, Ticker, Fecha y Valor
import re
informe = open('/content/drive/My Drive/ClassFiles/informe_acciones.txt', encoding='latin-1')
for linea in informe:
  linea = linea.rstrip()

  #Nombre de la accion y ticker
  nombre_ticker = re.findall(r'acción de .+', linea) #Busca el nombre de la acción y el ticker por separado busca el nombre completo de la acción con (.+) y el ticker sin espacios con (\S+) que está entre paréntesis
  if len(nombre_ticker) > 0:
    print('Nombre y ticker:', nombre_ticker)

  #Extraer fecha
  fecha = re.findall(r'^Fecha: .+', linea) # Busqueda de lineas que empiecen con Fecha: seguido de cualquier caracter (.) repetido una o mas veces (+) hasta el final de la linea
  if len(fecha) > 0:
      print('Fecha: ', fecha)

  #Extraer valor
  valor = re.findall(r'^Valor actual: .+', linea)
  if len(valor) > 0:
    print('Valor actual:', valor)

```

**3. Calcular y mostrar en pantalla:**

* La cantidad de datos extraídos.

* La sumatoria de los datos.

* El promedio de los datos.

* La varianza de los datos.

```python
archivo = open('/content/drive/MyDrive/ClassFiles/mbox.txt')

X_DSPAM_Confidence = []

for linea in archivo:
  linea = linea.rstrip()
  #ignorar lineas que no nos interesan
  if not linea.startswith('X-DSPAM-Confidence'):
    continue
    # Tomar los ultimo carcateres para extraer el valor de X-DSPAM-Confidence
  valor = float(linea[20:])
  X_DSPAM_Confidence.append(valor)

#Cantidad de datos
cantidad = len(X_DSPAM_Confidence)

#Sumatoria
suma = sum(X_DSPAM_Confidence)

#Promedio
promedio = suma / cantidad

#Varianza
suma_diferencias = 0
for x in X_DSPAM_Confidence:
  suma_diferencias += (x - promedio) ** 2
varianza = suma_diferencias / cantidad

print('Cantidad de datos: ', cantidad)
print('Sumatoria de los datos: ', suma)
print('Promedio de los datos: ', promedio)
print('Varianza de los datos: ', varianza)
```
